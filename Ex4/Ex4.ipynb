## 1. Read "NS_elevation.txt" into a NumPy array.

```python
import numpy as np
from osgeo import gdal
from matplotlib import pyplot as plt
from matplotlib import cm

"""
How to fully load a raster image into a NumPy array:
http://gis.stackexchange.com/questions/32995/how-to-fully-load-a-raster-into-a-numpy-array
"""

dataset = gdal.Open("NS_DEM_merged_high_compression.tif")

data = np.array(dataset.GetRasterBand(1).ReadAsArray())

data.shape
```

## 2. Get corner points and max / min values of data

```python
"""
Get UTM corner points from plotting
"""
data.ptp()
```

## 3. Plot array as an color scaled heatmap

```python
step = 1 #change this if you want to decimate the data

elevations= data[:4500:step,1700:10000:step]

plt.figure(figsize = [30,12], facecolor='white' )
plt.imshow(elevations, cmap = cm.gist_earth_r, vmin = 0, vmax=525)
plt.colorbar(shrink=0.75)
plt.show()
```

## 4. Do plot the elevation profileon a N-S line through the Old-Orchard Inn

```python
# North South line through Old-Orchard Inn
plt.figure(figsize = [1,5])
Orchard_segment = np.flipud(data[1950:3800,4100])
plt.plot(Orchard_segment, np.arange(len(Orchard_segment)) )
plt.axis('off')
plt.title('elevation profile from North to South')
plt.show()
```

```python
plt.figure(figsize = [30,12], facecolor='white' )
plt.imshow(elevations, cmap = cm.gist_earth_r, vmin = 0, vmax=525, alpha=1.0)

clipped = numpy.where(elevations <= 50, elevations, 0)

plt.imshow(clipped , cmap = cm.RdGy_r, vmin = -50, vmax=50, alpha = 0.5)

plt.colorbar(shrink=0.75)
plt.show()
```
